Mengisi missing value dengan nilai rata-rata(mean) tiap kolom

```
mean = wine_df['fixed_acidity'].mean()
```

```
mean
```

```
wine_df['fixed_acidity'] = wine_df['fixed_acidity'].fillna(mean)
```

```
mean_volatile_acidity = wine_df['volatile_acidity'].mean()
```

```
wine_df['volatile_acidity'] = wine_df['volatile_acidity'].fillna(mean_volatile_acidity)
```

```
mean_citric_acid = wine_df['citric_acid'].mean()
```

```
wine_df['citric_acid'] = wine_df['citric_acid'].fillna(mean_citric_acid)
```

```
mean_residual_sugar = wine_df['residual_sugar'].mean()
```

```
wine_df['residual_sugar'] = wine_df['residual_sugar'].fillna(mean_residual_sugar)
```

```
mean_chlorides = wine_df['chlorides'].mean()
```

```
wine_df['chlorides'] = wine_df['chlorides'].fillna(mean_chlorides)
```

```
mean_pH = wine_df['pH'].mean()
```

```
wine_df['pH'] = wine_df['pH'].fillna(mean_pH)
```

```
mean_sulphates = wine_df['sulphates'].mean()
```

```
wine_df['sulphates'] = wine_df['sulphates'].fillna(mean_sulphates)
```

```
from sklearn.tree import DecisionTreeClassifier
```

Karena menggunakan Algoritma Supervised Learning, maka kita pisahkan kolom yang menjadi input features dan target label

```
X = wine_df[['fixed_acidity','volatile_acidity','citric_acid','residual_sugar','chlorides','free_sulfur_dioxide','total_sulfur_dioxide','density','pH','sulphates','alcohol']]
y = wine_df['quality']
```

Membuat model

```
tree_model = DecisionTreeClassifier()
```

Melakukan model terhadap data

```
tree_model.fit(X,y)
```

Melakukan prediksi dengan menggunakan data yang ada di dalam dataset

```
tree_model.predict([[7.0, 0.27, 0.36, 20.7, 0.045, 45.0, 170.0, 1.0010, 3.00, 0.45, 8.8]])
```
